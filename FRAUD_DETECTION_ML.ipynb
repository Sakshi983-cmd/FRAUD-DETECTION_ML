```python
# Step 1: Importing Required Libraries
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import pickle
from google.colab import files

# Step 2: Upload the CSV file
uploaded = files.upload()  # This will prompt to upload the 'fraudTrain.csv' file

# Step 3: Load the dataset (after uploading)
train_df = pd.read_csv("fraudTrain.csv", nrows=50000)  # Make sure 'fraudTrain.csv' is uploaded

# Step 4: Features and Target
X_train = train_df.select_dtypes(include=['number']).drop('is_fraud', axis=1)  # Select numerical columns except the target
y_train = train_df['is_fraud']  # Target variable

# Step 5: Train the Random Forest Model
model = RandomForestClassifier(n_estimators=50, random_state=42)
model.fit(X_train, y_train)  # Train the model

# Step 6: Save the trained model using pickle
with open('/content/fraud_detection_model.pkl', 'wb') as file:
    pickle.dump(model, file)

# Step 7: Confirm that the model has been trained and saved
print("✅ Model trained and saved as 'fraud_detection_model.pkl'.")

# Step 8: Download the model pickle file
files.download('/content/fraud_detection_model.pkl')  # This will download the model file
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd
import joblib
from google.colab import files  #

# Upload test file (fraudTest.csv)
uploaded = files.upload()  #

# Load the test data
test_df = pd.read_csv("fraudTest.csv")

# Preprocess the test data
X_test = test_df.select_dtypes(include=['number']).drop('is_fraud', axis=1)
y_test = test_df['is_fraud']

# Split training data
X_train, X_test_split, y_train, y_test_split = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Train the model again on the split data
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test_split)

# Evaluation
print("Confusion Matrix:")
print(confusion_matrix(y_test_split, y_pred))

print("\nClassification Report:")
print(classification_report(y_test_split, y_pred))

# save the model
joblib.dump(model, 'fraud_detection_model.pkl')
print("Model saved as 'fraud_detection_model.pkl'")

# predictions on the new test data
y_test_pred = model.predict(X_test)

# Print predictions
print("\nPredictions on the uploaded test data:", y_test_pred)
```
